```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
matchMap = 'Bind'
matchId = 86365
playerId = 2829
playerName = 'GE skillZ'
opponentTeam = 'VLT'
startRound = 1
endRound = 12
defTeamId = []
plotTime = []
aR = []
```

```python
df = pd.read_csv('loc.csv')
df = df[(df.matchId == matchId) & (df.playerId == playerId) & (df.roundNumber >= startRound) & (df.roundNumber <= endRound)]
df.reset_index(inplace = True)
df = df[['roundNumber', 'playerId', 'roundTimeMillis', 'locationX','locationY', 'matchId']]
print(df)
#X = np.array(df['locationX'])
#Y = np.array(df['locationY'])
#plt.imshow(image)
#plt.scatter(X,Y)
```

```python
xloc = np.array(df['locationX'])
yloc = np.array(df['locationY'])
r= np.array(df['roundNumber'])
time = np.array(df['roundTimeMillis'])
#print(len(r))
x = 0
y = 0
z = 0
T = []
for i in range(startRound,max(r)+1):
    #print(i)
    df1 = (df.loc[df['roundNumber'] == i])
    #print(df1)
    y = len(df1)
    if(y == 1):
        T.append(time[x]) #append1
        x = x+y
        continue
    if(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteB,(xloc[x+1],yloc[x+1]),False) != -1) == False):
        T.append(time[x]) #append2
        x = x+y
        continue
    if(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteA,(xloc[x+1],yloc[x+1]),False) != -1) == False):
        T.append(time[x]) #append3
        x = x+y
        continue
    for j in range(x,x+y):
        #print(j)
        if(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteB,(xloc[j],yloc[j]),False) != -1) == False)):
            z = j-1
            break
        elif(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteA,(xloc[j],yloc[j]),False) != -1) == False)):
            z = j-1
            break    
        elif(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteA,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
            z = x+y-1
            break
        elif(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteB,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
            z = x+y-1
            break
    T.append(time[z])
        
    x = x+y
T = np.array(T)
T = (T/(1000))
#print(T.shape)
anchor = round(np.average(T),2)
print(T)
print(anchor)
```

```python
df2 = pd.read_csv('round.csv')
teamsId = np.unique(df2['attackingTeamId'])
df2 = df2[(df2.matchId == matchId) & (df2.roundNumber >= startRound) & (df2.roundNumber <= endRound)]
df2.reset_index(inplace = True)
df2 = df2[['roundNumber', 'matchId', 'winCondition', 'winningTeamId', 'spikePlant', 'spikePlantRoundTimeMillis','roundDurationMillis','attackingTeamId']]
print(df2)
atkTeam = np.unique(df2['attackingTeamId'])
for i in teamsId:
    if i != atkTeam:
        defTeamId.append(i)
print(defTeamId)
```

```python
roundTime = np.array(df2['roundDurationMillis'])
plantTime = np.array(df2['spikePlantRoundTimeMillis'])
spikePlant = np.array(df2['spikePlant'])
winningTeamId = np.array(df2['winningTeamId'])
#print(roundTime)
rT = ((roundTime/(1000)))
pT = ((plantTime/1000))

for i in range(len(df2)):
    if(np.isnan(pT[i])):
        plotTime.append(rT[i])
    else:
        plotTime.append(pT[i])

aR = T/plotTime

#print(T)
#print(pT)
#print(aR)
#print(T/rT)
for i in range(len(aR)):
    if aR[i] > 1:
        aR[i] = 1.0
print(aR)
print(spikePlant)
```

```python
y= np.arange(startRound,endRound+1,1)
fig = plt.figure(figsize = (18,9),facecolor = '#fff3d1')
ax = plt.axes()
ax.set_facecolor('#fff3d1')
plt.grid(color = '#a3a3a3', linestyle='--',linewidth=0.15)
plt.plot(y,plotTime, color = '#eb34c3', label = 'Round Objective Time')
plt.plot(y,T, label = 'Round Anchor Time')
plt.plot(y,rT,color= '#deb500', label = 'Total Round Time')
#plt.hlines(np.average(plotTime),xmin = startRound,xmax = endRound,color = '#eb34c3',linestyle = ':',label = 'Average Objective Time')
#plt.hlines(np.average(T),xmin = startRound,xmax = endRound,linestyle = ':',label = 'Average Anchor Time')
#plt.hlines(np.average(rT),xmin = startRound,xmax = endRound,color = '#deb500',linestyle = ':',label = 'Average Round Time')
for i in range(len(y)):
    if (winningTeamId[i] == defTeamId):
        if (spikePlant[i] != 'True'):
            plt.scatter(y[i],rT[i],color = 'green',s = 150,marker = '*', label = "Defender's Frag Win")
        else:
            plt.scatter(y[i],rT[i],color = 'green',s = 130,marker = 'o', label = "Defender's Objective Win")
    else:
        if (df2.spikePlant[i] != 'True'):
            plt.scatter(y[i],rT[i], color = 'red', s = 150, marker = '*', label = "Attacker's Frag Win")
        else:
            plt.scatter(y[i],rT[i], color = 'red', s = 130,marker = 'o', label = "Attacker's Objective Win")
    plt.annotate(round(aR[i],2), (y[i]-0.08, rT[i] + 2), size = 10)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(),fontsize = '10',loc = 'lower right')

plt.title(f'{playerName} || Site Anchor Plot || {matchMap} || vs {opponentTeam} \n [Avg Anchor Ratio: {np.average(aR):.2f}]',size = 18)
plt.xlabel('Round number ->',size = 16)
plt.ylabel('Time [s] ->',size = 16)


plt.savefig(f'{playerName}.png',dpi = 600,bbox_inches = 'tight',facecolor = 'auto')
```
